```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import random
from scipy.stats import linregress
```

```python
df = pd.read_csv('..//Data//HAM10000_metadata.csv')
clean_df = df.dropna(axis = 'rows')
clean_df = clean_df.loc[clean_df['sex'] != 'unknown']
clean_df = clean_df.loc[clean_df['localization'] != 'unknown']
clean_df
```

```python
grouped = clean_df['localization'].value_counts()

# grouped = clean_df.groupby('localization').count()[['dx']].sort_values(by=['dx'], ascending=False)
grouped
```

```python
grouped.plot(kind='bar')
plt.title("Skin Cancer on Body")
plt.xlabel("Area of Body")
plt.ylabel("Total Instances for Area")

plt.savefig("Images/Skin_Cancer_on_Body.png")

plt.show()

# Below bar chart shows the total instances for each area of the body that have been effected by skin cancer.
```

```python
bins = [0, 39.9 ,49.9, 59.9, 69.9, 79.9, 80]
group_names = ['Under Forty', 'Forties', 'Fifties', 'Sixties', 'Seventies', 'Over Eighty']

clean_df2 = clean_df
clean_df2['Age Groups'] = pd.cut(clean_df2['age'], bins, labels=group_names, include_lowest=True)
print(clean_df2)
```

```python
grouped_age = clean_df2['Age Groups'].value_counts().plot(kind="pie", labels=group_names, autopct='%1.1f%%', explode=[0.1,0,0,0,0,0], shadow=True, textprops={'fontsize': 10, 'color':"w", 'weight':'bold'})

fig = plt.gcf()
fig.set_size_inches(5,5)
fig.set_facecolor('#8B8B83')

plt.title("% of Age Groups Effected", fontsize= 14, fontweight='bold')
plt.axis('off')

plt.savefig("Images/Pct_of_Age_Groups_Effected.png")

plt.show()

# Below pie chart shows the age groups effected by skin cancer broken down by percentiles.
```

```python
color_map = {"abdomen": "#FF1493", "acral": "#458B74", "back": "#5F9EA0", "chest": "black", "ear": "blue", "face": "#FFD700", "foot": "#CD3333","genital": "#98F5FF", "hand": "#FF6103", "lower extremity": "#8B8378", "neck": "cornflowerblue", "scalp": "darkgray", "trunk": "darkorchid", "upper extremity": "#458B00"}
color_list = color_map.values()


clean_df2.groupby('localization')["Age Groups"].value_counts().unstack(0).plot.barh(figsize=(15,12),color=color_list)


plt.title("Localization vs. Age Group")
plt.xlabel("Count of People w/ Skin Cancer")
plt.ylabel("Age Groups")

plt.rcParams.update({'font.size': 40}, )


plt.savefig("Images/Localization_vs_Age_Group.png")

plt.show()

# Below horizontal bar chart shows the area of the body grouped by age decade. 
```
